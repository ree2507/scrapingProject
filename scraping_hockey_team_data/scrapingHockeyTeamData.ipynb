1. import library

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://www.scrapethissite.com/pages/forms/?page_num=1"
```

2. scraping satu halaman dan menampilkan nama

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

team_elements = soup.find_all('td', class_='name')

team_names = [element.get_text(strip=True) for element in team_elements]

print(team_names)
```

3. mengambil data lengkap dari banyak halaman

```python
def get_team_data_from_page(page_num):
    """Scrapes team data (name, year, wins, losses, win %) from a specific page number."""
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page_num}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    teams_data = []
    team_rows = soup.select('tr.team') # Select rows with team data

    for row in team_rows:
        name_element = row.select_one('td.name')
        year_element = row.select_one('td.year')
        wins_element = row.select_one('td.wins')
        losses_element = row.select_one('td.losses')
        win_pct_element = row.select_one('td.pct')

        team_data = {
            "Team Name": name_element.get_text(strip=True) if name_element else None,
            "Year": int(year_element.get_text(strip=True)) if year_element else None,
            "Wins": int(wins_element.get_text(strip=True)) if wins_element else None,
            "Losses": int(losses_element.get_text(strip=True)) if losses_element else None,
            "Win %": float(win_pct_element.get_text(strip=True)) if win_pct_element else None,
        }
        teams_data.append(team_data)
    return teams_data

all_teams_data = {}
for page_num in range(1, 26):
    teams_data = get_team_data_from_page(page_num)
    all_teams_data[f"Page {page_num}"] = teams_data
    print(f"Scraped {len(teams_data)} teams from page {page_num}")
```

4. menampilkan data yang dikumpulkan per halaman

```python
for page, teams in all_teams_data.items():
    print(f"\n{page}:")
    for team in teams:
        print(team)
```

5. mengimpor data ke csv

```python
# Flatten the dictionary of lists into a single list of dictionaries
all_teams_list = []
for page, teams in all_teams_data.items():
    all_teams_list.extend(teams)

# Create a pandas DataFrame
df_teams = pd.DataFrame(all_teams_list)

# Save the DataFrame to a CSV file
df_teams.to_csv("teams_data.csv", index=False)

print("Data berhasil disimpan ke dalam teams_data.csv")
```
